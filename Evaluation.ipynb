```
# Tải các thư viện cần thiết
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import r2_score as r2
# Cài đặt giá trị seed ngẫu nhiên
np.random.seed(2)
```

```
# Sinh dữ liệu ngẫu nhiên tượng trưng cho 100 khách hàng trong một cửa hàng
# và thói quen mua sắm của họ
x = np.random.normal(3, 1, 100)
y = np.random.normal(150, 40, 100) / x
```

```
# Toàn bộ dữ liệu
# x tượng trưng cho số phút trước khi quyết định mua
# y tượng trưng cho số tiền đã dùng sau khi mua
plt.scatter(x, y)
plt.show()
```

```
# Chia tập dữ liệu thành hai phần train:test theo tỷ lệ 0.8:0.2
# Lấy 80 dữ liệu đầu tiên (đếm từ 0 đến 79)
train_x = x[:80]
train_y = y[:80]
# Lấy 20 dữ liệu cuối cùng (đếm từ 80 đến hết)
test_x = x[80:]
test_y = y[80:]
```

```
plt.scatter(train_x, train_y)
plt.show()
```

```
plt.scatter(test_x, test_y)
plt.show()
```

```
# Vẽ đường cong hồi quy đa thức
model = np.poly1d(np.polyfit(train_x, train_y, 4))
line = np.linspace(0, 6, 100)

plt.scatter(train_x, train_y)
plt.plot(line, model(line))
plt.show()
```

Để tìm mối quan hệ giữa x và y, người ta dùng R2, quy ước 0 là không có quan hệ và 1 là quan hệ hoàn toàn. Phương thức `r2_score()` sẽ giúp tìm mối quan hệ đó.

```
# So sánh y thực sự và y dự đoán
score = r2(train_y, model(train_x))
print(score)
```

```
# So sánh trên dữ liệu test
score = r2(test_y, model(test_x))
print(score)
```
