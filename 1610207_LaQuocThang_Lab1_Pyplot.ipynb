```
import matplotlib.pyplot as plt
import numpy as np
```

## Vẽ đồ thị đường `plt.plot`

- Hàm `plot()` dùng để vẽ các điểm (points, markers) trong đồ thị.
- Mặc định, `plot()` vẽ đường thẳng.

```
# Tạo giá trị liên tục từ 0 đến 9, bước nhảy là 0.05
# Sau đó tính sin và cos từng phần tử
a = np.arange(0, 10, 0.05)
sin_a = np.sin(a)
cos_a = np.cos(a)
print(sin_a)
```

```
# Đưa giá trị vào
plt.plot(sin_a)
plt.plot(cos_a)
# Đặt tên trục tung và trục hoành và tên đồ thị
plt.ylabel('trục y')
plt.xlabel('trục x')
plt.title('đồ thị hình sin và cos')
# Vẽ
plt.show()
```

Để vẽ đồ thị với các điểm đánh dấu, sử dụng tham số *ký hiệu chuỗi phím tắt*, chẳng hạn 'o', '*',...

```
b = np.arange(0, 10, 0.2)
sin_b = np.sin(b)
cos_b = np.cos(b)
```

```
# Vẽ đồ thị bằng điểm tròn (o)
plt.plot(sin_b, 'ro') # red round
plt.plot(cos_b, 'g*') # green star
# [xmin, xmax, ymin, ymax]
plt.axis([0, 100, -1, 1])
plt.show()
```

```
c = np.arange(0., 5., 0.2)
# trục x, trục y, kiểu dáng nét đứt đỏ
plt.plot(c, c, 'r--')
# trục x^2, trục y và hình vuông xanh dương
plt.plot(c**2, c, 'bs')
# trục x, y^2 và tam giác xanh lá
plt.plot(c, c**2, 'g^')
plt.show()
```

## Vẽ đồ thị phân tán `plt.scatter`

- Đồ thị phân tán là đồ thị mà mỗi giá trị trong tập dữ liệu được đại diện bởi một chấm (dot).
- Để vẽ, sử dụng hàm `scatter()`

```
x = [5,7,8,7,2,17,2,9,4,11,12,9,6]
y = [99,86,87,88,111,86,103,87,94,78,77,85,86]
# Vẽ biểu đồ điểm (biểu đồ phân tán, scatter plot)
plt.scatter(x, y)
plt.show()
```

## Vẽ đa đồ thị `plt.subplots()`

Chúng ta có thể vẽ nhiều đồ thị trong một hình bằng hàm `subplots()`

```
#plot 1:
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])

# Vẽ đồ thị 1 dòng 2 cột
fig, axs = plt.subplots(1, 2)
axs[0].plot(x1, y1)
axs[0].set_title('Đồ thị 1')

#plot 2:
x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

axs[1].plot(x2, y2)
axs[1].set_title('Đồ thị 2')

plt.show()
```

```
x1 = np.array([0, 1, 2, 3])
y1 = np.array([3, 8, 1, 10])

# Vẽ đồ thị 3 dòng 1 cột
fig, axs = plt.subplots(3, 1)

axs[0].plot(x1, y1)
axs[0].set_title('Đồ thị 1')

x2 = np.array([0, 1, 2, 3])
y2 = np.array([10, 20, 30, 40])

axs[1].plot(x2, y2)
axs[1].set_title('Đồ thị 2')

x3 = np.array([0, 1, 2, 3])
y3 = np.array([1, 5, 3, 7])

axs[2].plot(x3, y3)
axs[2].set_title('Đồ thị 3')

plt.show()
```

```
def f(x):
    return np.exp(-x) * np.cos(2 * np.pi * x)

t1 = np.arange(0.0, 5.0, 0.1)
t2 = np.arange(0.0, 5.0, 0.02)

fig, axs = plt.subplots(1, 2)

axs[0].plot(t1, f(t1), 'bo', t2, f(t2), 'k')
axs[1].plot(t2, np.cos(2 * np.pi * t2), 'r--')

plt.show()
```

```
# Fixing random state for reproducibility
np.random.seed(19680801)

# make up some data in the open interval (0, 1)
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure()

# linear
plt.subplot(221)
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')
plt.grid(True)

# log
plt.subplot(222)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)

# symmetric log
plt.subplot(223)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthresh=0.01)
plt.title('symlog')
plt.grid(True)

# logit
plt.subplot(224)
plt.plot(x, y)
plt.yscale('logit')
plt.title('logit')
plt.grid(True)
# Adjust the subplot layout, because the logit one may take more space
# than usual, due to y-tick labels like "1 - 10^{-3}"
plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.25,
                    wspace=0.35)

plt.show()
```

### Biểu đồ tròn `plt.pie()`


```
# Nhãn dữ liệu
labels = 'Nông lâm - Thủy sản', 'Công nghiệp - Xây dựng', 'Dịch vụ'
# Dữ liệu, đơn vị nghìn tỷ đồng
data2000 = [108.4, 162.2, 171.1]
data2005 = [175.9, 344.2, 319]
data2010 = [407.6, 814.1, 759.2]
data2014 = [679, 1307.9, 1537.2]
# Tạo hiệu ứng "explode" cho phần đầu
explode = (0.1, 0, 0)

fig, axs = plt.subplots(2, 2, figsize = (10, 10))

# Năm 2000
# Tạo biểu đồ tròn cho ô đầu tiên, dữ liệu năm 2000, góc 90 độ, thêm % kích thước,
# và hiệu ứng explode
wedges, texts, autotexts = axs[0, 0].pie(data2000, startangle = 90, 
                                         autopct = '%1.1f%%', explode = explode)
axs[0, 0].axis('equal')
axs[0, 0].set_title('Năm 2000')
# Năm 2005
axs[0, 1].pie(data2005, startangle = 90, autopct = '%1.1f%%', explode = explode)
axs[0, 1].axis('equal')
axs[0, 1].set_title('Năm 2005')
# Năm 2010
axs[1, 0].pie(data2010, startangle = 90, autopct = '%1.1f%%', explode = explode)
axs[1, 0].axis('equal')
axs[1, 0].set_title('Năm 2010')
# Năm 2014
axs[1, 1].pie(data2014, startangle = 90, autopct = '%1.1f%%', explode = explode)
axs[1, 1].axis('equal')
axs[1, 1].set_title('Năm 2014')

plt.legend(wedges, labels, bbox_to_anchor=(2, 2))
plt.suptitle('Cơ cấu tổng sản phẩm trong nước (GDP) theo giá thực tế phân theo khu vực kinh tế, giai đoạn 2000 - 2014')
plt.show()
```

### Tô màu `plt.fill()`

- `fill()` vẽ một đa giác dựa trên danh sách các cạnh theo trục tọa độ *x*, *y*.
- Ví dụ sau vẽ một bông tuyết [Koch](https://en.wikipedia.org/wiki/Koch_snowflake).

```
"""
order: độ sâu đệ quy
scale: chiều dài cạnh của bông tuyết dựa trên tam giác
"""
def koch_snowflake(order, scale = 10):
    def _koch_snowflake_complex(order):
        if order == 0:
            # initial triangle
            angles = np.array([0, 120, 240]) + 90
            return scale / np.sqrt(3) * np.exp(np.deg2rad(angles) * 1j)
        else:
            ZR = 0.5 - 0.5j * np.sqrt(3) / 3

            p1 = _koch_snowflake_complex(order - 1)  # start points
            p2 = np.roll(p1, shift=-1)  # end points
            dp = p2 - p1  # connection vectors

            new_points = np.empty(len(p1) * 4, dtype=np.complex128)
            new_points[::4] = p1
            new_points[1::4] = p1 + dp / 3
            new_points[2::4] = p1 + dp * ZR
            new_points[3::4] = p1 + dp / 3 * 2
            return new_points
    
    points = _koch_snowflake_complex(order)
    x, y = points.real, points.imag
    # x và y là hai danh sách các tọa độ của bông tuyết Koch
    return x, y
```

```
x1, y1 = koch_snowflake(1)
x2, y2 = koch_snowflake(2)
x3, y3 = koch_snowflake(3)
x4, y4 = koch_snowflake(4)

fig, axs = plt.subplots(2, 2, figsize = (8, 8))
axs[0, 0].fill(x1, y1, facecolor = 'lightsalmon')
axs[0, 1].fill(x2, y2, facecolor = 'lightsalmon')
axs[1, 0].fill(x3, y3, facecolor = 'lightsalmon')
axs[1, 1].fill(x4, y4, facecolor = 'lightsalmon')
plt.show()
```
